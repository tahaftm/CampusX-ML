### Closed for equation method - OLS Technique <br>
##### Custom Linear Regression class

```python
class LinearRegression:
    def __init__(self):
        self.w = None
        self.b = None
    
    def fit(self,X,y):
        y_bar = y.mean()
        x_bar = X.mean()
        for idx in range(0,len(X)):
            Xi = X[idx]
            yi = y[idx]
            self.w = np.dot((yi - y_bar),(Xi - x_bar))/(Xi - x_bar)
            
        self.b = y_bar - np.dot(self.w,x_bar)
        print("W = ", self.w)
        print("b = ", self.b)
    def predict(self,X):
        return self.w * X + self.b
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
y_train.mean()
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test[0])
```
