```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
import matplotlib.pyplot as plt
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
plt.scatter(X,y)
```

### Finding b thorugh ols and then gradient descent algorithm

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lin_reg.predict(X), color = 'red')
```

Applying gradient descent assuming the value if slope is contant 78.35 

```python
y_pred = ((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lin_reg.predict(X), color = 'red', label = "OLS")
plt.plot(X,y_pred, color = 'green', label = "b = 0")
plt.legend()
```

```python
# initializing the values of b as 0
m = 78.35
b = 0

loss_slope = -2 * np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
```

```python
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lin_reg.predict(X), color = 'red', label = "OLS")
plt.plot(X,y_pred, color = '#A3E4D7', label = "b = 0")
plt.plot(X,y_pred1, color = '#00a65a', label = "b = {}".format(b))
plt.legend()
```

```python
loss_slope = -2 * np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1
step_size = loss_slope*lr
b = b - step_size
b
```

```python
y_pred2 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lin_reg.predict(X), color = 'red', label = "OLS")
plt.plot(X,y_pred, color = '#A3E4D7', label = "b = 0")
plt.plot(X,y_pred1, color = '#00a65a', label = "b = {}".format(b))
plt.plot(X,y_pred2, color = "#00ff8c", label = "b = {}".format(b))
plt.legend()
```

Writing everythin in a loop

```python
epochs = 10
b = -10
eta = 0.1

plt.scatter(X,y)

for epoch in range(epochs):
    loss_slope = -2 * np.sum(y-m*X.ravel() - b) 
    b = b - (eta*loss_slope)
    y_pred = ((78.35*X)+b).reshape(4)
    
    plt.plot(X,lin_reg.predict(X), color = 'red')
    plt.plot(X,y_pred, color = "#67E600")
    
```
