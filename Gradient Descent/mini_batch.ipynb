```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
X,y = load_diabetes(return_X_y=True)
print(X.shape)
print(y.shape)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
reg = LinearRegression()
reg.fit(X_train,y_train)
print(reg.coef_)
print(reg.intercept_)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
X_train.shape
```

```python
import random
class BGDRegressor:
    def __init__(self, batch_size,learning_rate = 0.01, epochs = 100):
        self.lr = learning_rate
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.batch_size = batch_size
    
    def fit(self,X,y):
        # initializing the coefficients
        self.intercept_ = 0
        m = X.shape[1]
        self.coef_ = np.ones(m)

        for i in range(self.epochs):
            for j in range(int(X_train.shape[0]/self.batch_size)):
                # update all the coefficents and the intercepts
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                y_hat = np.dot(X[idx],self.coef_)+self.intercept_
                intercept_der = -2 * np.mean(y[idx]-y_hat)
                self.intercept_ = self.intercept_ - (self.lr*intercept_der)
                
                coef_der = -2 * np.dot((y[idx] - y_hat), X[idx])
                self.coef_ = self.coef_ - (self.lr*coef_der)

    def predict(self,X):
        return np.dot(X,self.coef_)+self.intercept_
```

```python
bgd = BGDRegressor(epochs = 75, learning_rate=0.01,batch_size=int(X_train.shape[0]/50))
```

```python
bgd.fit(X_train, y_train)
```

```python
pred=bgd.predict(X_test)
pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```
